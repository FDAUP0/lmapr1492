<p>Dans ce Jupyter Notebook, nous allons télécharger le fichier CIF du matériau ScNiSb <code>mp-3432</code>. </p>
    
<p>Ensuite nous allons déterminer les 3 premiers pics d'intensité du diffractogramme correspondant à notre matériau en indiquant les indices hkl associés. Nous considérons un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å.
</p>

```python
#télécharger le fichier CIF du matériau ScNiSb 
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator

my_user_api_key = "RoIT1CX7DvB59O2S" # Materials Project legacy API key
m = MPRester(api_key=my_user_api_key)
structure = m.get_structure_by_material_id("mp-3432", conventional_unit_cell=True)

with open("ScNiSb.cif", "w") as f:
    f.write(structure.to(fmt="cif"))
```

```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator

xrd = XRDCalculator(1.54060)
pat = xrd.get_pattern(structure)
xrd.show_plot(structure,annotate_peaks=False,two_theta_range=[0,180])
```

```python
#Récupération des informations sur les 3 premiers pics : 

# angles theta : 
theta2 = pat.x

# Intensités : 
I = pat.y

# indices hkl associés à chacun des pics
hkl = pat.hkls

# Tri des différents pics selon l'intensité
combined_lists = list(zip(theta2, I, hkl))
sorted_lists = sorted(combined_lists, key=lambda x: x[1], reverse=True)
theta2, I, hkl = zip(*sorted_lists)

# print(sorted_I)
# print(sorted_theta2)
# print(sorted_hkl) 
```

```python
print('Premier pic de diffraction : ')
print('Angle theta : {} degrés'.format(theta2[0]/2))
print('Intensité normalisée : {} '.format(I[0]))
print('Indices hkl : {} '.format(hkl[0]))
```

```python
print('Premier pic de diffraction : ')
print('Angle theta : {} degrés'.format(theta2[1]/2))
print('Intensité normalisée : {} '.format(I[1]))
print('Indices hkl : {} '.format(hkl[1]))
```

```python
print('Troisième pic de diffraction : ')
print('Angle theta : {} degrés'.format(theta2[2]/2))
print('Intensité normalisée : {} '.format(I[2]))
print('Indices hkl : {} '.format(hkl[2]))
```
