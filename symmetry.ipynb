<h2>1) Symétrie de translation :</h2>
<p>La symétrie de translation déplace un point par un vecteur donné. Ici le vecteur vaut (1/2,1/2,0). La translation conserve la chiralité.</p>
<p> équation représentant la translation : 
\begin{equation}
\begin{pmatrix}
x \\
y \\
z
\end{pmatrix}
+
\begin{pmatrix}
1/2 \\
1/2 \\
0
\end{pmatrix}
=
\begin{pmatrix}
x +1/2 \\
y  + 1/2\\
z
\end{pmatrix}
\end{equation}</p>
<p>Nous prenons l'atome de Nickel au point (1/4, 3/4, 1/4) qui sera affecté par la translation en se placant dans les coordonnées (3/4, 5/4, 1/4). L'atome est placé hors de la maille après effet de la symmétrie, mais d'autres ne sont pas placé en dehors après effet de la symétrie.</p>

figure 2 : symétrie appliquée sur un atome de Nickel. On remarque qu'un autre atome de nickel est mis en évidence. Celui ci désigne la position de l'atome translaté dans sa maille.

<h2>2) Symétrie de Rotation d'ordre 2 :</h2>
<p>La symétrie de est une symétrie ponctuelle, qui s'applique autour d'un axe, ici l'axe (0, 0, 1). La rotation conserve la chiralité.</p>
<p> équation représentant l'effet de la symétrie : 
\begin{equation}
\begin{pmatrix}
-1 & 0 & 0 \\
0 & -1 & 0 \\
0 & 0 & 1
\end{pmatrix}
\begin{pmatrix}
x \\
y \\
z
\end{pmatrix}
=
\begin{pmatrix}
-x \\
-y \\
z
\end{pmatrix}
\end{equation}</p>
</p>Nous remarquons que le déterminant de la matrice de rotation vaut bien 1, signe que la rotation conserve la chiralité.</p>
<p>Nous prenons l'atome d'Antimoine (Sb) au point (1/2, 1/2, 1) qui sera affecté par la rotation en se placant dans les coordonnées (-1/2, -1/2, 1).</p>


figure 3 : effet de symétrie montré sur la maille vide (pour plus de clarté)


figure 4 : symétrie appliquée sur l'atome d'Antimoine (1/2, 1/2, 1). L'origine est en bas à droite. On remarque le signe de la rotation rouge autour de son axe, mais aussi le vecteur rouge représentant également la rotation.

<h2>3) Symétrie de Roto-inversion d'ordre 4 :</h2>
<p>La symétrie de roto-inversion est ici une inversion suivie d'une rotation. C'est une symétrie ponctuelle, qui s'applique autour d'un point, ici l'axe (0, 1/2, 0). La roto-inversion inverse la chiralité.</p>
<p> équation qui applique la symétrie : 
\begin{equation}
\begin{pmatrix}
-1 & 0 & 0 \\
0 & 0 & 1 \\
0 & -1 & 0
\end{pmatrix}
\begin{pmatrix}
x \\
y \\
z
\end{pmatrix}
+
\begin{pmatrix}
0 \\
\frac{1}{2} \\
\frac{1}{2}
\end{pmatrix}
=
\begin{pmatrix}
-x \\
z + \frac{1}{2} \\
-y + \frac{1}{2}
\end{pmatrix}
\end{equation}</p>
<p>Nous prenons l'atome de Scandium (Sc) au point (1/2, 1/2, 1/2) qui sera affecté par la symétrie en se placant dans les coordonnées (-1/2, 1, 0).</p>


figure 5 : effet de symétrie montré sur la maille vide (pour plus de clarté)


figure 6 : symétrie appliquée sur l'atome de Scandium. Le point (0, 0, 0) est en haut à gauche. On remarque le vecteur bleu clair symbolisant l'inversion et les vecteurs rouges symbolisant la rotation.
